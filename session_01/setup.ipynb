#### Bonus: NumPy for NLP - Bag of words example

BUNU TEKRAR KONTROL EDECEĞİM

```python
# Create a simple vocabulary
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are pets"
]

# Build vocabulary
vocab = set()
for sentence in corpus:
    for word in sentence.split():
        vocab.add(word)

vocab = sorted(vocab)
word_to_idx = {word: i for i, word in enumerate(vocab)}
print(f"Vocabulary size: {len(vocab)}")
print(f"Vocabulary: {vocab}\n")

# Create bag-of-words vectors
def create_bow(sentence, vocab, word_to_idx):
    bow = np.zeros(len(vocab))
    for word in sentence.split():
        if word in word_to_idx:
            bow[word_to_idx[word]] += 1
    return bow

# Create BoW for first sentence
sentence1 = corpus[0]
bow_vector = create_bow(sentence1, vocab, word_to_idx)

print(f"Sentence: '{sentence1}'")
print(f"BoW vector: {bow_vector}")
print(f"Vector shape: {bow_vector.shape}")

# Create document-term matrix
dt_matrix = np.array([create_bow(sent, vocab, word_to_idx) for sent in corpus])
print(f"\nDocument-Term Matrix shape: {dt_matrix.shape}")
print("Document-Term Matrix:\n", dt_matrix)
```

```python
# Create tensor with requires_grad=True to track computations
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1  # y = x^2 + 3x + 1

# Compute gradient dy/dx
y.backward()

print(f"x = {x}")
print(f"y = {y}")
print(f"dy/dx at x=2: {x.grad}")  # Should be 2*2 + 3 = 7
```

#### Simple linear regression example and PyTorch Training Loop

```python
# Generate synthetic data
torch.manual_seed(42)
np.random.seed(42)

X = torch.linspace(0, 10, 100).reshape(-1, 1)
true_w = 2.0
true_b = 1.0
y = true_w * X + true_b + 0.5 * torch.randn(X.shape)

# Define a simple linear model
class LinearModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = torch.nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearModel()

# Loss and optimizer
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training loop
print("Training PyTorch model...")
for epoch in range(100):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")

# Learned parameters
learned_w = model.linear.weight.item()
learned_b = model.linear.bias.item()
print(f"\nTrue: w={true_w}, b={true_b}")
print(f"Learned: w={learned_w:.4f}, b={learned_b:.4f}")
```

#### Working with text in PyTorch (NLP teaser)

```python
# Simple vocabulary
words = ["hello", "world", "nlp", "pytorch", "fun"]
word_to_ix = {word: i for i, word in enumerate(words)}
ix_to_word = {i: word for word, i in word_to_ix.items()}

# Convert sentence to indices
sentence = ["hello", "nlp", "fun"]
indices = [word_to_ix[w] for w in sentence]
print(f"Sentence indices: {indices}")

# One-hot encoding (simplified)
def one_hot(indices, vocab_size):
    return torch.eye(vocab_size)[indices]

one_hot_vectors = one_hot(indices, len(words))
print(f"One-hot vectors:\n{one_hot_vectors}")

# Embedding layer (learnable representations)
embedding = torch.nn.Embedding(num_embeddings=len(words), embedding_dim=5)
embedded = embedding(torch.tensor(indices))
print(f"\nEmbedded shape: {embedded.shape}")
print(f"Embedded vectors:\n{embedded}")
```

#### Exercise 3 - Simple text classifier stub

```python
# TODO: Create a tiny neural network for binary sentiment classification
# Input: 10-dimensional word embedding, Output: positive/negative (2 classes)

class TinySentimentClassifier(torch.nn.Module):
    def __init__(self, embedding_dim=10, hidden_dim=5, output_dim=2):
        super().__init__()
        # Define layers here
        self.fc1 = torch.nn.Linear(embedding_dim, hidden_dim)
        self.fc2 = torch.nn.Linear(hidden_dim, output_dim)
        self.relu = torch.nn.ReLU()

    def forward(self, x):
        # Define forward pass
        x = self.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Test with random input
model = TinySentimentClassifier()
sample_input = torch.rand(3, 10)  # batch of 3 samples, 10-dim each
output = model(sample_input)
print(f"Output shape: {output.shape}")  # Should be [3, 2]
```

#### Bonus Exercise - NumPy vs PyTorch performance

```python
# Compare matrix multiplication speed (CPU only)
import time

size = 1000
np_a = np.random.rand(size, size)
np_b = np.random.rand(size, size)
torch_a = torch.tensor(np_a)
torch_b = torch.tensor(np_b)

# NumPy
start = time.time()
np_result = np_a @ np_b
np_time = time.time() - start

# PyTorch
start = time.time()
torch_result = torch_a @ torch_b
torch_time = time.time() - start

print(f"NumPy time: {np_time:.4f} seconds")
print(f"PyTorch time: {torch_time:.4f} seconds")
print(f"PyTorch is {np_time/torch_time:.2f}x faster")
```

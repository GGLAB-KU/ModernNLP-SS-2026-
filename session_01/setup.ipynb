#### Bonus: NumPy for NLP - Bag of words example

```python
# Create a simple vocabulary
corpus = [
    "the cat sat on the mat",
    "the dog sat on the log",
    "cats and dogs are pets"
]

# Build vocabulary
vocab = set()
for sentence in corpus:
    for word in sentence.split():
        vocab.add(word)

vocab = sorted(vocab)
word_to_idx = {word: i for i, word in enumerate(vocab)}
print(f"Vocabulary size: {len(vocab)}")
print(f"Vocabulary: {vocab}\n")

# Create bag-of-words vectors
def create_bow(sentence, vocab, word_to_idx):
    bow = np.zeros(len(vocab))
    for word in sentence.split():
        if word in word_to_idx:
            bow[word_to_idx[word]] += 1
    return bow

# Create BoW for first sentence
sentence1 = corpus[0]
bow_vector = create_bow(sentence1, vocab, word_to_idx)

print(f"Sentence: '{sentence1}'")
print(f"BoW vector: {bow_vector}")
print(f"Vector shape: {bow_vector.shape}")

# Create document-term matrix
dt_matrix = np.array([create_bow(sent, vocab, word_to_idx) for sent in corpus])
print(f"\nDocument-Term Matrix shape: {dt_matrix.shape}")
print("Document-Term Matrix:\n", dt_matrix)
```

#### Gradients and autograd (NumPy doesn't have this!)

#### Differantiation Operations with PyTorch
<img src="visuals/auto_diff.png" width="900">

```python
# Tensors that track operations for gradient computation
x = torch.tensor(2.0, requires_grad=True)
print(f"x = {x}, requires_grad={x.requires_grad}")

# Define a function y = x^2 + 3x + 1
y = x**2 + 3*x + 1
print(f"y = {y}")

# Compute gradient dy/dx
y.backward()
print(f"dy/dx at x=2: {x.grad}")  # Should be 2*2 + 3 = 7

# More complex example
x = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
y = (x ** 2).sum()  # y = 1^2 + 2^2 + 3^2 = 14
y.backward()
print(f"\nx = {x}")
print(f"y = {y}")
print(f"Gradients dy/dx: {x.grad}")  # Should be [2, 4, 6]

# Detaching tensors (stop gradient tracking)
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2
z = y.detach()  # z doesn't require gradient
print(f"\ny requires grad: {y.requires_grad}")
print(f"z requires grad: {z.requires_grad}")

# Using with torch.no_grad() context
with torch.no_grad():
    w = x * 3
    print(f"w requires grad (inside no_grad): {w.requires_grad}")
```

```python
# Create tensor with requires_grad=True to track computations
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3 * x + 1  # y = x^2 + 3x + 1

# Compute gradient dy/dx
y.backward()

print(f"x = {x}")
print(f"y = {y}")
print(f"dy/dx at x=2: {x.grad}")  # Should be 2*2 + 3 = 7
```

#### Simple linear regression example and PyTorch Training Loop

```python
# Generate synthetic data
torch.manual_seed(42)
np.random.seed(42)

X = torch.linspace(0, 10, 100).reshape(-1, 1)
true_w = 2.0
true_b = 1.0
y = true_w * X + true_b + 0.5 * torch.randn(X.shape)

# Define a simple linear model
class LinearModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = torch.nn.Linear(1, 1)

    def forward(self, x):
        return self.linear(x)

model = LinearModel()

# Loss and optimizer
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# Training loop
print("Training PyTorch model...")
for epoch in range(100):
    y_pred = model(X)
    loss = loss_fn(y_pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")

# Learned parameters
learned_w = model.linear.weight.item()
learned_b = model.linear.bias.item()
print(f"\nTrue: w={true_w}, b={true_b}")
print(f"Learned: w={learned_w:.4f}, b={learned_b:.4f}")
```

## Custom Dataset for NLP

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np

# Simple text classification dataset
class SimpleTextDataset(Dataset):
    """A simple dataset for text classification."""

    def __init__(self, texts, labels, vocab=None, max_length=10):
        self.texts = texts
        self.labels = labels
        self.max_length = max_length

        # Build vocabulary if not provided
        if vocab is None:
            self.vocab = self.build_vocab(texts)
        else:
            self.vocab = vocab

        # Special tokens
        self.pad_idx = 0  # Padding token index
        self.unk_idx = 1  # Unknown token index

    def build_vocab(self, texts):
        """Build vocabulary from texts."""
        word_counts = {}
        for text in texts:
            for word in text.lower().split():
                word_counts[word] = word_counts.get(word, 0) + 1

        # Create vocab: 0=PAD, 1=UNK, then other words
        vocab = {'<PAD>': 0, '<UNK>': 1}
        for word in word_counts:
            vocab[word] = len(vocab)
        return vocab

    def text_to_indices(self, text):
        """Convert text to list of indices."""
        indices = []
        for word in text.lower().split():
            idx = self.vocab.get(word, self.unk_idx)
            indices.append(idx)

        # Pad or truncate to max_length
        if len(indices) > self.max_length:
            indices = indices[:self.max_length]
        else:
            indices = indices + [self.pad_idx] * (self.max_length - len(indices))

        return torch.tensor(indices, dtype=torch.long)

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        label = self.labels[idx]

        # Convert text to indices
        indices = self.text_to_indices(text)

        return {
            'input_ids': indices,
            'label': torch.tensor(label, dtype=torch.long),
            'text': text  # Keep original text for debugging
        }

# Create sample data
texts = [
    "I love this movie",
    "This film is terrible",
    "Great acting and story",
    "Waste of time and money",
    "Absolutely fantastic",
    "Boring and predictable"
]

labels = [1, 0, 1, 0, 1, 0]  # 1=positive, 0=negative

# Create dataset
dataset = SimpleTextDataset(texts, labels, max_length=5)
print(f"Dataset size: {len(dataset)}")
print(f"Vocabulary size: {len(dataset.vocab)}")
print(f"Vocabulary: {dataset.vocab}")

# Access a sample
sample = dataset[0]
print(f"\nSample 0:")
print(f"  Text: {sample['text']}")
print(f"  Input IDs: {sample['input_ids']}")
print(f"  Label: {sample['label']}")
```

```python
# Create DataLoader for batching
batch_size = 2
dataloader = DataLoader(
    dataset,
    batch_size=batch_size,
    shuffle=True,  # Shuffle for training
    num_workers=0  # Set >0 for multiprocessing
)

print(f"Number of batches: {len(dataloader)}")

# Iterate through batches
for batch_idx, batch in enumerate(dataloader):
    print(f"\nBatch {batch_idx + 1}:")
    print(f"  Input IDs shape: {batch['input_ids'].shape}")
    print(f"  Labels: {batch['label']}")
    print(f"  Texts: {batch['text']}")

    if batch_idx == 1:  # Show first 2 batches only
        break

# Different DataLoader configurations
print("\nDataLoader configurations:")
print(f"Batch size: {batch_size}")
print(f"Shuffle: {dataloader.dataset is not None}")
print(f"Number of workers: {dataloader.num_workers}")
```

```python
class SimpleClassifier(nn.Module):
    """A simple neural network for text classification."""

    def __init__(self, vocab_size, embedding_dim, hidden_dim, num_classes):
        super(SimpleClassifier, self).__init__()

        # Embedding layer: converts word indices to dense vectors
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)

        # Hidden layer
        self.fc1 = nn.Linear(embedding_dim, hidden_dim)

        # Output layer
        self.fc2 = nn.Linear(hidden_dim, num_classes)

        # Activation functions
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        # x shape: (batch_size, sequence_length)

        # Embedding: (batch_size, seq_len, embedding_dim)
        embedded = self.embedding(x)

        # Average over sequence length (simple pooling)
        # This creates a document representation
        pooled = embedded.mean(dim=1)  # (batch_size, embedding_dim)

        # Hidden layer
        hidden = self.relu(self.fc1(pooled))
        hidden = self.dropout(hidden)

        # Output layer (logits)
        output = self.fc2(hidden)

        return output

# Model parameters
vocab_size = len(dataset.vocab)
embedding_dim = 32
hidden_dim = 64
num_classes = 2  # Binary classification

# Create model
model = SimpleClassifier(vocab_size, embedding_dim, hidden_dim, num_classes)
print("Model architecture:")
print(model)
print(f"\nTotal parameters: {sum(p.numel() for p in model.parameters())}")

# Move model to device
device = torch.device('cuda' if torch.cuda.is_available() else
                     'mps' if torch.backends.mps.is_available() else
                     'cpu')
model = model.to(device)
print(f"Model device: {next(model.parameters()).device}")
```

```python
# Test forward pass with a batch
model.eval()  # Set to evaluation mode
sample_batch = next(iter(dataloader))
input_ids = sample_batch['input_ids'].to(device)

with torch.no_grad():
    logits = model(input_ids)
    probabilities = torch.softmax(logits, dim=1)
    predictions = torch.argmax(logits, dim=1)

print("Forward pass test:")
print(f"Input shape: {input_ids.shape}")
print(f"Logits shape: {logits.shape}")
print(f"Logits:\n{logits}")
print(f"\nProbabilities:\n{probabilities}")
print(f"\nPredictions: {predictions}")
print(f"True labels: {sample_batch['label']}")
```

```python
# Hyperparameters
learning_rate = 0.01
num_epochs = 50

# Loss function
criterion = nn.CrossEntropyLoss()

# Optimizer
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# For tracking metrics
train_losses = []
train_accuracies = []

print("Training Setup:")
print(f"Learning rate: {learning_rate}")
print(f"Number of epochs: {num_epochs}")
print(f"Loss function: {criterion.__class__.__name__}")
print(f"Optimizer: {optimizer.__class__.__name__}")
```

```python
# Training function
def train_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    correct = 0
    total = 0

    for batch in dataloader:
        # Move data to device
        input_ids = batch['input_ids'].to(device)
        labels = batch['label'].to(device)

        # Forward pass
        optimizer.zero_grad()
        outputs = model(input_ids)
        loss = criterion(outputs, labels)

        # Backward pass
        loss.backward()
        optimizer.step()

        # Statistics
        total_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    avg_loss = total_loss / len(dataloader)
    accuracy = 100 * correct / total

    return avg_loss, accuracy

# Evaluation function
def evaluate(model, dataloader, criterion, device):
    model.eval()
    total_loss = 0
    correct = 0
    total = 0

    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(device)
            labels = batch['label'].to(device)

            outputs = model(input_ids)
            loss = criterion(outputs, labels)

            total_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    avg_loss = total_loss / len(dataloader)
    accuracy = 100 * correct / total

    return avg_loss, accuracy

# Training loop
print("Starting training...")
print("-" * 60)

for epoch in range(num_epochs):
    train_loss, train_acc = train_epoch(model, dataloader, criterion, optimizer, device)
    train_losses.append(train_loss)
    train_accuracies.append(train_acc)

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}] "
              f"Loss: {train_loss:.4f}, "
              f"Accuracy: {train_acc:.2f}%")

print("-" * 60)
print("Training completed!")
```

```python
def predict_text(model, text, dataset, device):
    """Predict sentiment for a single text."""
    model.eval()

    # Convert text to indices
    indices = dataset.text_to_indices(text)
    input_ids = indices.unsqueeze(0).to(device)  # Add batch dimension

    with torch.no_grad():
        logits = model(input_ids)
        probabilities = torch.softmax(logits, dim=1)
        prediction = torch.argmax(logits, dim=1)

    sentiment = "Positive" if prediction.item() == 1 else "Negative"
    confidence = probabilities[0][prediction].item()

    return sentiment, confidence, probabilities[0].cpu().numpy()

# Test on new texts
test_texts = [
    "This movie was amazing and wonderful",
    "Terrible film, completely boring",
    "Pretty good, I enjoyed it",
    "Not great, could be better"
]

print("Predictions on new texts:")
print("-" * 60)

for text in test_texts:
    sentiment, confidence, probs = predict_text(model, text, dataset, device)
    print(f"Text: '{text}'")
    print(f"Sentiment: {sentiment} (confidence: {confidence:.2%})")
    print(f"Probabilities: [Negative: {probs[0]:.2%}, Positive: {probs[1]:.2%}]")
    print()
```
